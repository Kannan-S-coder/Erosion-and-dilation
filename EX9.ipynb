```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_img():
    blank_img =np.zeros((900,900))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img,text='KANNAN S',org=(50,500), fontFace=font,fontScale= 5,color=(255,255,255),thickness=25,lineType=cv2.LINE_AA)
    return blank_img
```

```python
def display_img(img):
    plt.imshow(img,cmap='gray')
    plt.show()
```

```python
img = load_img()
```

```python
display_img(img)
```

## Erosion

```python
kernel = np.ones((5,5),dtype=np.uint8)
```

```python
kernel
```

```python
img=load_img()
erosion1=cv2.erode(img,kernel,iterations=1)
```

```python
display_img(erosion1)
```

## Dilation

```python
dilation=cv2.dilate(img,kernel,iterations=1)
```

```python
display_img(dilation)
```
